```python
def standardise_product_name(name):
    name = name.split("'")
    name = name[1][1:].strip()
                                        #return name.split("'")[1][1:].strip().title()
    temp = name.split('-')
    if len(temp) > 1:
        name = temp[0].capitalize() + '-' + temp[1]
    else:
        name = name.title()
    return name

st['snack'] = st['snack'].apply(standardise_product_name)
st
```

```python
df = pd.merge(vm,st, on='snack', how='left')
df = df.set_index('machine_code')
df
```

```python
def auth():
    attempt = 0
    status = False
    def show_msg(msg, attempt):
        attempt += 1
        print(f'{msg}, current attempts are {attempt}')
        return attempt
    while attempt < 4:
        username = input("Please enter username:>> ")
        if username == "admin":
            status = True
            while status == True and attempt < 4:
                try:
                    password = int(input("Please enter password:>> "))
                    if password < 10:
                        print("Login successful...")
                        return True
                    else:
                        attmept = show_msg("Invalid password", attempt)
                except:
                    attempt = show_msg('Password must be a number', attempt)
        else:
            attempt = show_msg('Invalid username', attempt)
    return False
                    


def add_product():
    status = False
    print("This is the current stock of the vending machine:\n\n", df2)
    select = input("Select a slot to add one more product:").upper()
    if select in df2.index.values:
        if df2.loc[select].isnull().any():
            print("this slot is empty, try another one")
        else:
            while status == False:
                print('Enter Quantity')
                try:
                    qty = int(input(" >> " ))
                    if qty > 0:
                        new_qty = df2.loc[select, 'amount'] + qty
                        if new_qty > 8:
                            print('total quantity cannot exceed 8, try another one')

                        else:
                            df2.loc[select, 'amount'] = new_qty
                        print("the product has been added, here is the updated stock:", df2)
                        status = True
                    else:
                        print('quantity cannot be less than 1...')
                except:
                    print('Invalid input')

    else:
        print ("Invalid slot, try again")


def buy_product():
    status = False
    print("This is the current stock of the vending machine:\n\n", df2)
    select = input("Select a product to buy:").upper()
    if select in df2.index.values:
        if df2.loc[select].isnull().any():
            print('No products left on this slot, try another one')
        else:
            while status == False:
                print('Product to buy:', df2.loc[select, 'snack'])
                print('Enter Quantity')
                try:
                    qty = int(input(" >> " ))
                    if qty > 0 and qty < df.loc[select, 'price']:
                        new_qty = df.loc[select, 'price'] - qty
                        cost = df.loc[select, 'price'] * qty
                        print('Price:', cost)
                        pay = int(input("Please enter the amount of money to pay: "))
                        if pay < 0:
                            print("you inserted 'negative' money! try again!")
                        elif pay < cost:
                            print("you need to pay more, try again")
                        else:
                            change = pay - cost
                            print("Your change is: ", change)
                            df2.loc[select, 'amount'] = new_qty
                            print(f"The number of snacks left in {select} is {new_qty}")
                            status = True
                except:
                    print('quantity cannot be less than 1 or more than available stock...')    
    else:
        print("Invalid slot, try again") 
    


def menu():
    print('1. Login in (admin only)')
    print('2. Add products (admin only)')
    print('3. Buy products')
    print('4. Exit')
    return

def vending_machine():
    status = False
    print('Welcome to vending machine. ')
    while True:
        
        menu()
        option = int(input('Select an option: '))

        if option == 1:
            if status == False:
                if auth() == True:
                    status = True
            else:
                print('You already logged in, back to main menu')
        elif option == 2:
            if status == True:
                add_product()
            else:
                print("login first")
        elif option == 3:
            buy_product()
        elif option == 4:
            print('Thank you for your patronage..')
            break
        else:
            print('Invalid option')
            break
            
            
if __name__ == "__main__":
    vending_machine()
```
